```python
import pandas as pd
dataset = pd.read_csv('amazon_reviews.csv')
dataset.info()
dataset.head(25)
```

```python
useful_Dataset = dataset.loc[:,['overall','reviewText']]
useful_Dataset.head(10)
```

```python
def sentiment(x):
    if x >= 4 :
        return 'positive'
    else :
        return 'negative'

useful_Dataset['sentiment'] = useful_Dataset['overall'].apply(sentiment)
print(useful_Dataset)
useful_Dataset['sentiment'].value_counts()
```

```python
import numpy as np
useful_Dataset['reviewText'].fillna('')
useful_Dataset['reviewText'].replace(['','NaN',' ',np.nan,'Null','missing'],pd.NA)
useful_Dataset['reviewText'].dropna(inplace=True)
useful_Dataset.drop(index = 125,inplace = True)
useful_Dataset['reviewText'].isnull().sum()
```

```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

import nltk
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    tokens = word_tokenize(text)
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return ' '.join(tokens)

useful_Dataset['reviewText'] = useful_Dataset['reviewText'].apply(preprocess_text)
print(useful_Dataset.head())
```

```python
sentiment_mapping = {'positive': 1, 'negative': 0}
useful_Dataset['sentiment'] = useful_Dataset['sentiment'].map(sentiment_mapping)

print(useful_Dataset['sentiment'].value_counts())
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(useful_Dataset['reviewText'], useful_Dataset['sentiment'], test_size=0.3, random_state=42)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer(num_words=10000) 
tokenizer.fit_on_texts(x_train)

X_train_seq = tokenizer.texts_to_sequences(x_train)
X_test_seq = tokenizer.texts_to_sequences(x_test)

max_length = 100
X_train_padded = pad_sequences(X_train_seq, maxlen=max_length, padding='post', truncating='post')
X_test_padded = pad_sequences(X_test_seq, maxlen=max_length, padding='post', truncating='post')

print("Padded training data shape:", X_train_padded.shape)
print("Padded testing data shape:", X_test_padded.shape)
```

```python
from tensorflow.keras.utils import to_categorical

num_classes = 2  
y_train_encoded = to_categorical(y_train, num_classes=num_classes)
y_test_encoded = to_categorical(y_test, num_classes=num_classes)

print("Encoded training labels shape:", y_train_encoded.shape)
print("Encoded testing labels shape:", y_test_encoded.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GRU, Dense
from tensorflow.keras.optimizers import Adam

model = Sequential([
    Embedding(input_dim=10000, output_dim=128, input_length=max_length),  
    GRU(64, return_sequences=False),  
    Dense(num_classes, activation='softmax')  
])

# Compile the model
model.compile(optimizer=Adam(0.01), loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(
    X_train_padded, y_train_encoded,
    validation_data=(X_test_padded, y_test_encoded),
    epochs=10, batch_size=64
)
```

```python
loss, accuracy = model.evaluate(X_test_padded, y_test_encoded)
print("Test Accuracy:", accuracy)
```

```python
def predict_sentiment(review):
    sequence = tokenizer.texts_to_sequences([review])
    padded_sequence = pad_sequences(sequence, maxlen=max_length, padding='post', truncating='post')

    prediction = model.predict(padded_sequence)

    predicted_class_index = np.argmax(prediction, axis=1)[0]

    sentiment_mapping = {0: "negative", 1: "positive"}
    sentiment = sentiment_mapping.get(predicted_class_index, "unknown")

    return sentiment
```

```python
while True:
  new_review = input("Enter a movie review (or type 'exit' to quit): ")
  if new_review.lower() == "exit":
    break
  sentiment = predict_sentiment(new_review)
  print("Sentiment of the review is:", sentiment)
```
